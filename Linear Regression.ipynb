```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=np.array([1,2,3,4,5])
```

```python
y=np.array([3,4,2,4,5])
```

```python
plt.xlabel("value of X")
plt.ylabel("value of Y")
plt.title("Liner regresstion")
plt.plot(x,y,color="red",marker="D",label="prediction of y")
plt.legend()
```

```python
# n1=length of x
# n2-length of y
n1=x.size
n2=y.size
```

```python
# finding mean of X and Y
mean_x=np.mean(x)
mean_y=np.mean(y)
```

```python
# finding slope m
upper=0
lower=0
j=0
for i in x:
    upper=upper+((i-mean_x)*(y[j]-mean_y))
    j+=1
    lower=lower+((i-mean_x)**2)

#print(upper)
#print(lower)
m=upper/lower
```

```python
# finding intercept c
c=mean_y-(m*mean_x)
```

```python
# finding y'
y2=(m*x)+c
```

```python
## final plotting of the y'
plt.xlabel("value of X")
plt.ylabel("value of Y")
plt.title("Liner regresstion")
plt.plot(x,y2,label="Line of regression")
plt.scatter(x,y,color="orange",label="scattered point")
plt.legend()
```
